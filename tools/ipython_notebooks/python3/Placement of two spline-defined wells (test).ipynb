```python
%matplotlib inline
import LogReader as lr
import LogDataStructures as ls
reader = lr.LogReader("/home/einar/Documents/GitHub/PCG/fieldopt_output/adgprs4/")
cc = ls.CaseContainer(reader.case_log, reader.property_uuid_name_map)
```

# 1 Base Case Data

```python
base_case = cc.evaluated_cases[0]
print('Timestamp: ', base_case.timestamp)
print('UUID: ', base_case.uuid)
print('Objective function value: ', base_case.objective_function_value)
print('Number of vars: ', len(base_case.variables))
print('Variable names and values:')
for var in base_case.variables:
    print(var.name, '\t', var.value)
```

## 1.1 Well Data

```python
print('Number of wells in initial case: ', len(base_case.wells))
print('Well names: ', ', '.join(base_case.well_names))
print('Well information:')
for well in base_case.wells:
    print('\t', well.name)
    print('\tHeel: ', well.heel)
    print('\tToe: ', well.toe)
```

# 2 Best Case Data

```python
o = ls.Optimizer(reader.optimization_log, cc)
best_case = o.best_case_pr_iteration[-1]
print('Timestamp: ', best_case.timestamp)
print('UUID: ', best_case.uuid)
print('Objective function value: ', best_case.objective_function_value)
print('Number of vars: ', len(best_case.variables))
print('Variable names and values:')
for var in best_case.variables:
    print(var.name, '\t', var.value)
```

## 2.1 Well Data

```python
print('Number of wells in initial case: ', len(best_case.wells))
print('Well names: ', ', '.join(best_case.well_names))
print('Well information:')
for well in best_case.wells:
    print('\t', well.name)
    print('\tHeel: ', well.heel)
    print('\tToe: ', well.toe)
```

# 3 Objective Function

```python
iterations = [i+1 for i in range(len(o.best_case_id_pr_iteration))]
obj_fun_vals = []
for c in o.best_case_pr_iteration:
    obj_fun_vals.append(c.objective_function_value)

print('Iterations: ', iterations)    
print(obj_fun_vals)
for i in range(len(iterations)):
    print('Iteration ', iterations[i], ': ', obj_fun_vals[i])
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
fig.set_size_inches(10,6)
ax = fig.add_subplot(111)
ax.plot(iterations, obj_fun_vals)
fig.show()
```

# 4 Well Path Plots

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def set_3daxis_limits(axis):
    axis.set_xlim(0,1920)
    axis.set_ylim(0,1920)
    axis.set_zlim(1700, 1724)
```

## 4.1 Base Case and Tentative Best Case

```python
fig = plt.figure()
fig.set_size_inches(10,6)
ax = fig.add_subplot(111, projection='3d')
ax.set_title('Base Case')
injector = base_case.get_well("INJECTOR")
producer = base_case.get_well("PRODUCER")
injector_xs = [injector.heel[0], injector.toe[0]]
injector_ys = [injector.heel[1], injector.toe[1]]
injector_zs = [injector.heel[2], injector.toe[2]]
producer_xs = [producer.heel[0], producer.toe[0]]
producer_ys = [producer.heel[1], producer.toe[1]]
producer_zs = [producer.heel[2], producer.toe[2]]
ax.plot(injector_xs, injector_ys, injector_zs, color='b')
ax.plot(producer_xs, injector_ys, producer_zs, color='r')
set_3daxis_limits(ax)
fig.show()

fig_best_case = plt.figure()
fig_best_case.set_size_inches(10,6)
ax_best_case = fig_best_case.add_subplot(111, projection='3d')
ax_best_case.set_title('Tentative Best Case')
injector = best_case.get_well("INJECTOR")
producer = best_case.get_well("PRODUCER")
injector_xs = [injector.heel[0], injector.toe[0]]
injector_ys = [injector.heel[1], injector.toe[1]]
injector_zs = [injector.heel[2], injector.toe[2]]
producer_xs = [producer.heel[0], producer.toe[0]]
producer_ys = [producer.heel[1], producer.toe[1]]
producer_zs = [producer.heel[2], producer.toe[2]]
ax_best_case.plot(injector_xs, injector_ys, injector_zs, color='b')
ax_best_case.plot(producer_xs, injector_ys, producer_zs, color='r')
set_3daxis_limits(ax_best_case)
fig_best_case.show()
```

## 4.2 Last Simulated Case

```python
last_simulated_case = cc.evaluated_cases[-1]
fig_last_simulated_case = plt.figure()
fig_last_simulated_case.set_size_inches(10,6)
ax_last_simulated_case = fig_last_simulated_case.add_subplot(111, projection='3d')
ax_last_simulated_case.set_title('Last Simulated Case')
injector = last_simulated_case.get_well("INJECTOR")
producer = last_simulated_case.get_well("PRODUCER")
injector_xs = [injector.heel[0], injector.toe[0]]
injector_ys = [injector.heel[1], injector.toe[1]]
injector_zs = [injector.heel[2], injector.toe[2]]
producer_xs = [producer.heel[0], producer.toe[0]]
producer_ys = [producer.heel[1], producer.toe[1]]
producer_zs = [producer.heel[2], producer.toe[2]]
ax_last_simulated_case.plot(injector_xs, injector_ys, injector_zs, color='b')
ax_last_simulated_case.plot(producer_xs, injector_ys, producer_zs, color='r')
set_3daxis_limits(ax_last_simulated_case)
fig_best_case.show()
```

## 4.3 Tentative Best Case Projections

```python
injector = best_case.get_well("INJECTOR")
producer = best_case.get_well("PRODUCER")
injector_xs = [injector.heel[0], injector.toe[0]]
injector_ys = [injector.heel[1], injector.toe[1]]
injector_zs = [injector.heel[2], injector.toe[2]]
producer_xs = [producer.heel[0], producer.toe[0]]
producer_ys = [producer.heel[1], producer.toe[1]]
producer_zs = [producer.heel[2], producer.toe[2]]

f, (ax_xy, ax_xz, ax_yz) = plt.subplots(1,3)
f.set_size_inches(10,6)
f.tight_layout(pad=0.4, w_pad=2.5, h_pad=1.0)

# XY
ax_xy.plot(injector_xs, injector_ys, color='b')
ax_xy.plot(producer_xs, producer_ys, color='r')
ax_xy.set_title('XY-projection')
ax_xy.set_xlabel('x')
ax_xy.set_ylabel('y')
ax_xy.set_xlim(0,1920)
ax_xy.set_ylim(0,1920)

# XZ
ax_xz.plot(injector_xs, injector_zs, color='b')
ax_xz.plot(producer_xs, producer_zs, color='r')
ax_xz.set_title('XZ-projection')
ax_xz.set_xlabel('x')
ax_xz.set_ylabel('z')
ax_xz.set_xlim(0,1920)
ax_xz.set_ylim(1700,1724)

# YZ
ax_yz.plot(injector_ys, injector_zs, color='b')
ax_yz.plot(producer_ys, producer_zs, color='r')
ax_yz.set_title('YZ-projection')
ax_yz.set_xlabel('y')
ax_yz.set_ylabel('z')
ax_yz.set_xlim(0,1920)
ax_yz.set_ylim(1700,1724)

f.show()
```
